```
import pandas as pd
import numpy as np
import math
import random
import operator
from sklearn.preprocessing import MinMaxScaler
```

```
df = pd.read_csv("http://jain-harshil.github.io/pdf/CC%20GENERAL.csv")
df = df.fillna(0)
mms = MinMaxScaler()
mms.fit(df.loc[:,df.columns != 'CUST_ID'])
df = mms.transform(df.loc[:,df.columns != 'CUST_ID'])
df = pd.DataFrame(df)
```

```
def distance(point,l):
  minimum = math.inf
  for i in range (len(l)):
    sub = list(map(operator.sub, point, l[i]))
    distance = np.linalg.norm(np.array(sub))
    if distance < minimum:
      minimum = distance
  return minimum
```

```
b = []
for i in range (len(df)):
  b.append(list(df.iloc[i][1:]))
```

```
def greedy_k_center(a,k):
  final = []
  counter = 0
  i = random.randrange(len(df))
  final.append(a[i])
  a.remove(a[i])
  counter+=1
  while (counter!=k):
    maximum = 0
    index = 0
    for i in range (len(a)):
      dist = distance(a[i],final)
      if dist>maximum:
        maximum = dist
        index = i
    final.append(a[index])
    a.remove(a[index])
    counter = counter + 1
  return final
```

```
for k in [2,4,8,20]:
  print("Points and the objective function evaluated at k =",k)
  a = []
  for i in range (len(df)):
    a.append(list(df.iloc[i][1:]))
  li = greedy_k_center(a,k)
  print(li)
  obj = 0
  df2 = np.array(df)
  i = []
  for k in range (len(li)):
    i.append(b.index(li[k]))
  for j in range (len(df2)):
    if j not in i:
      min_dist = float('inf')
      for centre in i:
        curr_dist = np.linalg.norm(df2[centre,:]-df2[j,:])
        if curr_dist < min_dist:
          min_dist = curr_dist
      obj = max(obj,min_dist)
  print(obj)
```

```
from itertools import combinations 
```

```
def optimal_k_centres(m,k):
  nCk = list(combinations([i for i in range (len(m))],k))
  min_dist = float('inf')
  li = []
  for i in nCk:
    o = 0
    for j in range (len(m)):
      if j not in i:
        mind = float('inf')
        for centre in i:
          curr_d = np.linalg.norm(m[centre,:]-m[j,:])
          if curr_d < mind:
            mind = curr_d
        o = max(o,mind)
    if o<min_dist:
      min_dist = o
      li.append(i)
    z = li[-1]
  return min_dist,z
```

```
df1 = np.array(df)
# The optimal centres have been calculated only for first 100 datapoints
loss,list1 = optimal_k_centres(df1[0:100,:],2)
print(loss)
b = []
for i in range (len(list1)):
  b.append(list(df1[i]))
print(*b)
```

```
df1 = np.array(df)

loss,list1 = optimal_k_centres(df1[0:100,:],3)
print(loss)
b = []
for i in range (len(list1)):
  b.append(list(df1[i]))
print(*b)
```

```
final = []
finalloss = []
while(len(final)<=60):
  a = []
  for i in range (len(df)):
    a.append(list(df.iloc[i][1:]))
  li = greedy_k_center(a,20)
  final.append(li)
  obj = 0
  df2 = np.array(df)
  i = []
  for k in range (len(li)):
    i.append(b.index(li[k]))
  for j in range (len(df2)):
    if j not in i:
      min_dist = float('inf')
      for centre in i:
        curr_dist = np.linalg.norm(df2[centre,:]-df2[j,:])
        if curr_dist < min_dist:
          min_dist = curr_dist
      obj = max(obj,min_dist)
  finalloss.append(obj)
```

```
print("The centers for k = 20 are: ")
f = finalloss.index(min(finalloss))
print(final[f])
print(finalloss[f])
```
