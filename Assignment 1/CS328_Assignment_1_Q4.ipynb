```
import pandas as pd

population = pd.read_csv('http://jain-harshil.github.io/pdf/population.csv')
gini = pd.read_csv('http://jain-harshil.github.io/pdf/gini.csv')
X = pd.DataFrame()
Y = pd.DataFrame()
final = pd.DataFrame(columns=['country','year','population_growth','gini_coefficient'])
z = list(population.columns)
countries = list(population['country'])
znew = z[30:59]
znew.insert(0,'country')
gininew = gini[znew]
z.remove('country')
z = list(z)
populationew = population[znew]
countries = list(populationew['country'])
znew.remove('country')
for i in range (194):
    for j in range (29):
        final = final.append({'country':countries[i],'year':znew[j],'population_growth':populationew.iloc[i][j+1],'gini_coefficient':gininew.iloc[i][j+1]},ignore_index = True)
```
